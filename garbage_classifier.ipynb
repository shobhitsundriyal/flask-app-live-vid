```python
import os 

base_path = ''
train = 'training/'
val = 'validation/'

classes = ['garbage', 'non_garbage']
batch_size = 32
```

```python
from keras.utils import np_utils
from keras.layers import Dense, Activation, Flatten
from keras.layers import merge, Input
from keras.models import Model
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.applications import VGG16
from keras.applications import imagenet_utils
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from keras.utils import np_utils
from imutils import paths
import numpy as np
import pickle
import random
import os

image_input = Input(shape=(300, 300, 3))
model = VGG16(input_tensor=image_input, weights='imagenet', include_top=False)
le = None
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input

data_dir_list = os.listdir(train)
image_data_list = []
```

```python
for dataset in data_dir_list:
    img_list=os.listdir(train+'/'+ dataset)
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    for img in img_list:
        img_path = train + '/'+ dataset + '/'+ img
        img = image.load_img(img_path, target_size=(300, 300))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        x = x/255
        print('Input image shape:', x.shape)
        image_data_list.append(x)
        
```

```python
img_data = np.array(image_data_list)
print (img_data.shape)
```

```python
img_data=np.rollaxis(img_data,1,0)
print (img_data.shape)
img_data=img_data[0]
print (img_data.shape)
```

```python
#ze = img_data[:1500]
#ze.shape
img_data = img_data[:1850]
```

```python
num_classes = 2
num_of_samples = img_data.shape[0]
labels = np.ones((num_of_samples,),dtype='int64')
```

```python
labels[0:908] = 0
labels[908:] = 1
print (np.count_nonzero(labels))
```

```python
Y = np_utils.to_categorical(labels, num_classes)
```

```python
#Shuffle the dataset
x,y = shuffle(img_data,Y, random_state=2)
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
np.unique(y_train), np.count_nonzero(y_train)
y_train
```

```python
model.summary()
```

```python
last_layer = model.get_layer('block5_pool').output
```

```python
x = Flatten(name='flatten')(last_layer)
```

```python
x = Dense(128, activation='relu', name='fc1')(x)
x = Dense(128, activation='relu', name='fc2')(x)
out = Dense(num_classes, activation='softmax', name='output')(x)
```

```python
m_model = Model(image_input, out)
m_model.summary()
```

```python
for layer in m_model.layers[:-3]:
    layer.trainable = False

m_model.summary()

m_model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
import time
```

```python
t=time.time()

hist = m_model.fit(X_train, y_train, batch_size=32, epochs=10, verbose=1, validation_data=(X_test, y_test))
print('Training time: %s' % (t - time.time()))
(loss, accuracy) = m_model.evaluate(X_test, y_test, batch_size=10, verbose=1)

print("[INFO] loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
```
